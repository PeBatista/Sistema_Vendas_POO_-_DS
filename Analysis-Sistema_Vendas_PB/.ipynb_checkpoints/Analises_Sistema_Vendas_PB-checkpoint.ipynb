# Estrutura de Dados do Sistema de Vendas

A estrutura de dados do sistema de vendas consiste em várias tabelas que armazenam informações sobre clientes, funcionários, compras, produtos e fornecedores. Abaixo, estão detalhadas as tabelas principais do sistema:

## Tabela `t_cliente`

Esta tabela armazena informações dos clientes do sistema.

| Coluna                        | Tipo           | Descrição                             |
|-------------------------------|----------------|---------------------------------------|
| cd_cliente                    | NUMBER(6)      | Código do cliente físico               |
| nm_cliente                    | VARCHAR2(100)  | Nome do cliente físico                 |
| ds_tipo_cliente_virtual       | VARCHAR2(100)  | Tipo de pessoa do cliente virtual      |
| ds_rg_cliente                 | VARCHAR2(20)   | RG do cliente                          |
| ds_senha_cliente_virtual      | VARCHAR2(30)   | Senha do cliente virtual               |
| ds_cpf_cnpj_cliente           | VARCHAR2(20)   | CPF ou CNPJ do cliente virtual          |
| dt_nascimento_cliente_virtual | DATE           | Data de nascimento do cliente virtual  |
| fl_cli_fis_vir                | CHAR(1)        | 0:Fisica e 1:Virtual |

## Tabela `t_compra`

Esta tabela registra as compras realizadas pelos clientes.

| Coluna           | Tipo        | Descrição                               |
|------------------|-------------|-----------------------------------------|
| cd_compra        | NUMBER(6)   | Código da compra física                  |
| cd_funcionario   | NUMBER(6)   | Código do funcionário responsável        |
| cd_cliente       | NUMBER(6)   | Código do cliente                        |
| dt_hora_compra   | DATE        | Data e hora da compra                    |
| sg_status_entrega| CHAR(1)     | Status de entrega da compra              |
| dt_data_entrega  | DATE        | Data de entrega do produto               |
| ob_observacao    | VARCHAR2(100)| Observação da compra                      |
| vl_desconto      | NUMBER(8,2) | Valor do desconto aplicado               |
| vl_frete         | NUMBER(8,2) | Valor do frete                           |
| fl_compra_fis_vir| CHAR(1)     | Indica se a compra é física ou virtual   |

## Tabela `t_produto_vendido`

Esta tabela relaciona os produtos vendidos em cada compra.

| Coluna      | Tipo      | Descrição                              |
|-------------|-----------|----------------------------------------|
| cd_produto  | NUMBER(10)| Código do produto vendido               |
| cd_compra   | NUMBER(6) | Código da compra física                  |
| ds_quantidade| NUMBER(5) | Quantidade do produto vendido fisicamente |

## Tabela `t_fornecedor`

Esta tabela armazena informações dos fornecedores.

| Coluna              | Tipo          | Descrição                              |
|---------------------|---------------|----------------------------------------|
| cd_fornecedor       | NUMBER(6)     | Código do fornecedor                    |
| nm_fornecedor       | VARCHAR2(100) | Nome do fornecedor                      |
| ds_cpf_fornecedor   | VARCHAR2(14)  | CPF do fornecedor                       |
| ds_rg_fornecedor    | VARCHAR2(20)  | RG do fornecedor                        |
| sg_genero_fornecedor| CHAR(1)       | Gênero do fornecedor                    |
| dt_cadastro_fornecedor | DATE        | Data de cadastro do fornecedor          |

## Tabela `t_funcionario`

Esta tabela armazena informações dos funcionários do sistema.

| Coluna               | Tipo          | Descrição                              |
|----------------------|---------------|----------------------------------------|
| cd_funcionario       | NUMBER(6)     | Código do funcionário                   |
| nm_funcionario       | VARCHAR2(100) | Nome do funcionário                     |
| ds_cargo_funcionario | VARCHAR2(100) | Cargo do funcionário                    |
| ds_cpf_funcionario   | VARCHAR2(14)  | CPF do funcionário                      |
| ds_rg_funcionario    | VARCHAR2(20)  | RG do funcionário                       |
| dt_nascimento        | DATE          | Data de nascimento do funcionário        |
| dt_cadastro          | DATE          | Data de cadastro do funcionário          |
| sg_genero_funcionario| CHAR(1)       | Gênero do funcionário                   |

## Tabela `t_produto`

Esta tabela armazena informações dos produtos disponíveis.

| Coluna         | Tipo        | Descrição                              |
|----------------|-------------|----------------------------------------|
| cd_produto     | NUMBER(10)  | Código do produto                       |

## Chaves Estrangeiras

As tabelas possuem as seguintes chaves estrangeiras:

- A tabela `t_contato_cliente_virtual` possui a chave estrangeira `fk_cli_cli_virt` referenciando a coluna `cd_cliente` da tabela `t_cliente`.
- A tabela `t_compra` possui a chave estrangeira `fk_cliente_compra` referenciando a coluna `cd_cliente` da tabela `t_cliente`, e a chave estrangeira `fk_funcio_compra` referenciando a coluna `cd_funcionario` da tabela `t_funcionario`.
- A tabela `t_produto_vendido` possui a chave estrangeira `fk_comp_fisi_prod` referenciando a coluna `cd_compra` da tabela `t_compra`, e a chave estrangeira `fk_prod_comp_fisiv1` referenciando a coluna `cd_produto` da tabela `t_produto`.
- A tabela `t_contato_funcionario` possui a chave estrangeira `fk_contato_funcionario` referenciando a coluna `cd_funcionario` da tabela `t_funcionario`.
- A tabela `t_produto` possui a chave estrangeira `fk_forn_prod` referenciando a coluna `cd_fornecedor` da tabela `t_fornecedor`.

# Flags Utilizadas

## fl_cli_fis_vir (Cliente Físico ou Virtual)

Esta flag indica se o cliente é físico ou virtual.

- Valor 1: Indica que o cliente é físico.
- Valor 0: Indica que o cliente é virtual.

## fl_compra_fis_vir (Compra Física ou Virtual)

Esta flag indica se a compra foi feita de forma física ou virtual.

- Valor 1: Indica que a compra foi feita fisicamente.
- Valor 0: Indica que a compra foi feita virtualmente.

###### Começo primeiramente com uma análise feita por regressão, onde se encontra na mesma pasta do Projeto - importo o modelo e peço de uma forma visual (com formulário, dados de teste para você testar este meu modelo originado de outra pasta)



```python
with open('modelo.pkl', 'rb') as arquivo:
    modelo = pickle.load(arquivo)

# Função para realizar a previsão e exibir o resultado
def fazer_previsao(b):
    vl_frete = float(vl_frete_input.value)
    ds_quantidade = float(ds_quantidade_input.value)
    vl_preco = float(vl_preco_input.value)
    nr_estoque = float(nr_estoque_input.value)

    entrada = [[vl_frete, ds_quantidade, vl_preco, nr_estoque]]
    resultado = modelo.predict(entrada)[0].round(2)
    print(f'O resultado da previsão é: {resultado} reais')

# Interface do formulário
vl_frete_input = widgets.Text(value='teste', description='Valor do frete:')
ds_quantidade_input = widgets.Text(value='12', description='Quantidade:')
vl_preco_input = widgets.Text(value='100', description='Preço:')
nr_estoque_input = widgets.Text(value='90', description='Estoque:')

submit_button = widgets.Button(description='Prever')
submit_button.on_click(fazer_previsao)

display(vl_frete_input, ds_quantidade_input, vl_preco_input, nr_estoque_input, submit_button)
```

# Tratagens

```python
compras_produtos_vendidos = compras.merge(produtos_vendidos,on='cd_compra',how='inner')
compras_produtos_vendidos_produtos = compras_produtos_vendidos.merge(produtos,on='cd_produto',how='inner')
clientes_contatos = clientes.merge(contatos_cliente_virtual,on='cd_cliente',how='left')
# também quero para os códigos de clientes, onde não tem atrelação, ou seja trazendo como None para este
dados = compras_produtos_vendidos_produtos.merge(clientes_contatos,on='cd_cliente',how='inner')
# preciso da FLAG de compra Física ou Virtual vinda da tabela de cliente, pois é ela em que posso confiar p dizer isso
dados['vl_total'] = ((dados['vl_preco'] * dados['ds_quantidade']) - dados['vl_desconto'])
```

```python
dados = dados.drop(columns=['fl_compra_fis_vir'])
# colocar como Minas Gerais, já que a localidade real, de minha loja é em São Paulo
dados['ds_estado_cliente_virtual'] = dados['ds_estado_cliente_virtual'].fillna('Minas Gerais')
```
